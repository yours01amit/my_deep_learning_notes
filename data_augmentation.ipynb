```python
from keras.preprocessing import image
from tensorflow.keras.preprocessing.image import ImageDataGenerator
import matplotlib.pyplot as plt
import numpy as np
```

```python
img = image.load_img('/content/cat3.jpg', target_size=(400, 400))
```

```python
img
```

```python
type(img)
```

```python
dataget = ImageDataGenerator(
    rotation_range=40,
    width_shift_range=0.2,
    height_shift_range=0.2,
    shear_range=0.2,
    zoom_range=0.2,
    horizontal_flip=True,
)
```

```python
img = image.img_to_array(img)
```

```python
img.shape
```

```python
type(img)
```

```python
input_batch = img.reshape(1,400,400,3)
```

```python
i = 0
for output in dataget.flow(input_batch, batch_size=1, save_to_dir='aug', save_prefix='cat', save_format='jpeg'):
  i += 1
  if i == 10:
    break
```
